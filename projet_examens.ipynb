```python
import xml.etree.ElementTree as ET
```

```python
# telechargement des données xml 
url1 = "https://www.w3schools.com/xml/plant_catalog.xml"
response = requests.get(url1)
data1 = response.text
data1
root = ET.fromstring(data1)
```

```python
plant_data = []
    
for plant in root.findall("PLANT"):
    common = plant.find("COMMON").text
    price = plant.find("PRICE").text.strip("$")  # Supprimer le symbole "$"
    plant_data.append({"COMMON": common, "PRICE": float(price)})  # Stocke PRICE en float

# Convertir la liste en DataFrame Pandas
plant_data = pd.DataFrame(plant_data)

# Afficher le DataFrame des plantes
print("DataFrame des plantes :")
print(plant_data.head())
```

```python
# affichons les plantes donc le prix est supperieur à 5
print("\nliste des plantes donc le prix est supperieur à 5")
for plant in root.findall('PLANT'):
    price_text = plant.find("PRICE").text.strip("$")  # Supprimer le symbole "$"
    price = float(price_text)  # Convertir en nombre flottant
    if price > 5.0:
        common = plant.find("COMMON").text
        print(f"- {common} (Price: ${price:.2f})")
```

```python
# liste contenant classer par zone
plant_by_zonne = {}
for plant in root.findall('PLANT'):
    zone = plant.find('ZONE').text
    common = plant.find('COMMON').text
    if zone in plant_by_zonne:
        plant_by_zonne[zone].append(common)
    else:
        plant_by_zonne[zone] = [common]

# affichage des plantes classer par zone
print("\nliste des plantes classer par zone")
# for zone, plantes in plant_by_zonne.items():
#     print(f"Zone {zone}:")
#     for plante in plantes:
#         print(f"- {plante}")

for zone, plants in plant_by_zonne.items():
    print(f"Zone {zone}: {', '.join(plants)}")
```
